```python
from selenium import webdriver
from bs4 import BeautifulSoup
import requests
import pandas as pd
from selenium.webdriver.common.keys import Keys

driver = webdriver.Chrome()
driver.get('https://www.naukri.com/')

job_keyword = 'Data Analyst'

driver.find_element_by_xpath('//*[@id="qsb-keyword-sugg"]').send_keys(job_keyword)
driver.find_element_by_xpath('//*[@id="root"]/div[3]/div[2]/section/div/form/div[3]/button').click()

driver.implicitly_wait(30)
```

```python
soup = BeautifulSoup(driver.page_source, 'lxml')
```

```python
job_cards = soup.find_all('article', class_ = 'jobTuple bgWhite br4 mb-8')
print(len(job_cards))
jobs = []
```

```python
from selenium.webdriver.chrome.options import Options

for job in job_cards:
    
    skills_list = []
    
    for skill in job.find('ul',class_ = 'tags has-description').find_all('li'):
        skills_list.append(skill.text)
    
    skills = str(', '.join(skills_list))
        
    url = job.find('a', class_ = 'title fw500 ellipsis')['href']
    options = Options()
    options.headless = True
    url_driver = webdriver.Chrome(options=options)
    url_driver.get(url)
    s = BeautifulSoup(url_driver.page_source, 'lxml')
    url_driver.close() 
    
    points = []  
    try:
        description_string = s.find('div', class_ = 'dang-inner-html').text
    except:
        description_string = 'No description available'
    
    info = {'position' : job.find('a', class_ = 'title fw500 ellipsis').text,
            'company' : job.find('a', class_ = 'subTitle ellipsis fleft').text,
           'Experience' : job.find('li', class_ = 'fleft grey-text br2 placeHolderLi experience').find('span').text,
           'salary' : job.find('li', class_ = 'fleft grey-text br2 placeHolderLi salary').find('span').text,
           'Location' : job.find('li', class_ = 'fleft grey-text br2 placeHolderLi location').find('span').text,
           'Skills' : skills,
           'Description' : description_string}
    jobs.append(info)
```

```python
job_list = pd.DataFrame(jobs)
job_list.drop_duplicates(inplace=True)
job_list
```
